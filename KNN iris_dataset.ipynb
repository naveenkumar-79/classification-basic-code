```python
import numpy as np
import pandas as pd
import sklearn
```

```python
df=pd.read_csv('/content/Iris.csv')
df=df.drop(['Id'],axis=1)
df
```

```python
df['Species'].value_counts()
```

```python
df['Species']=df['Species'].map({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
```

```python
print(df)
```

```python
df.sample(7)
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
training_data=pd.DataFrame()
training_data=x_train.copy()
training_data['y_train']=y_train
print(training_data)
testing_data=pd.DataFrame()
testing_data=x_test.copy()
testing_data['y_test']=y_test
print(testing_data)
```

```python
from sklearn.neighbors import KNeighborsClassifier
reg=KNeighborsClassifier(n_neighbors=5)
reg.fit(x_train,y_train)
```

```python
y_train_pred=reg.predict(x_train)
training_data['y_train_pred']=y_train_pred
training_data
```

```python
y_test_pred=reg.predict(x_test)
testing_data['y_test_pred']=y_test_pred
print(testing_data)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
print(confusion_matrix(y_train,y_train_pred))
print("accuracy of training data",accuracy_score(y_train,y_train_pred))
```

```python
confusion_matrix(y_test,y_test_pred)
print(confusion_matrix(y_test,y_test_pred))
print("accuracy of testing data",accuracy_score(y_test,y_test_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_test_pred))
```
